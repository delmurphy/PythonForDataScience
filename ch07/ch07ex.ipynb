```python
#ex7.7 Challenge project: reimplement NumPy array Output
#write a function that reimplements NumPy’s array formatting for two-dimensional arrays using loops. Assume the array contains only positive integer values.
#(NumPy outputs two-dimensional arrays in a nice column-based format that right-aligns every element in a field width.
#The field width’s size is determined by the array element value that requires the most character positions to display.)

def nparray(int_from, int_to, nrow, ncol):
    x = [i for i in range(int_from, int_to)]
    maxwidth = len(str(max(x)))
    for r in range(nrow):
        for c in range(ncol):
            print(f'{x[r*ncol + c]:>{maxwidth}}', end = ' ')
        print('')
                


nparray(0, 12, 3, 4)
```

```python
#ex7.7 Challenge project: reimplement DataFrame Output
#write a function that reimplements pandas DataFrame formatting using loops. Assume the DataFrame contains only positive integer values and row and column names are ints beginning wtih 0.
#(pandas displays DataFrames in a nice column-based format with row and column labels that right-aligns every element in a column based on the columns widest value.)

def pddf(vals, nrows, ncols):
    #get max width of values in vals for formatting:
    maxwidth = len(str(max(vals)))
    #print column labels
    print(' '*maxwidth, end = ' ')
    for col in range(ncols):
        print(f'{col:>{maxwidth}}', end = ' ')
    print() #print a new line after the col labels
    #print row values preceded by row label
    for row in range(nrows):
        print(f'{row:>{maxwidth}}', end = ' ')
        for col in range(ncols):
            print(f'{vals[row*ncols + col]:>{maxwidth}}', end = ' ')
        print()


pddf([i for i in range(99, 111)], 3, 4)
```
